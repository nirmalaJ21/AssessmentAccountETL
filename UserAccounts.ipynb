# QUESTION 1: Extract
A GET request to the following endpoint provides you with randomly generated user information: https://randomuser.me/api/

Make 25 requests to the endpoint and add each response into a pandas DataFrame.
Your resulting DataFrame should be 25 rows and 34 columns.
Take advantage of pandas’ json_normalize() function
Each request should have at least a one second wait before the next request


```python
# extract — QUESTION 1

# write your solution(s) here:
# Initialize an empty DataFrame
df = pd.DataFrame()

# Make 25 requests
for _ in range(25):
    # Send a GET request to the endpoint
    response = requests.get("https://randomuser.me/api/")
    
    # Wait for one second
    time.sleep(1)
    
    # Extract the JSON data from the response
    data = response.json()
    
    # Normalize the JSON data and append it to the DataFrame
    df = df.append(pd.json_normalize(data['results']))

# Reset the DataFrame index
df = df.reset_index(drop=True)

# Display the resulting DataFrame
print(df.shape)  # Output: (25, 34)
print(df)
```

# QUESTION 2:  Transform
Using the DataFrame from Question 1:
Create the following columns:
na_user: boolean representing if the user is based in North America (defined as: US, Mexico, or Canada)
age_at_registration: the user’s age when they registered their account
Update the phone and cell column to remove any characters except for a leading + symbol
Your DataFrame should be 25 rows and 36 columns


```python
# transform — QUESTION 2

# write your solution(s) here:
# QUESTION 2: Transform
north_american_countries = ['US', 'Mexico', 'Canada']
df['na_user'] = df['nat'].apply(lambda x: x in north_american_countries)

# Create age_at_registration column
df['age_at_registration'] = pd.to_datetime(df['registered.date']).dt.year - pd.to_datetime(df['dob.date']).dt.year

# Update phone and cell columns
df['phone'] = df['phone'].str.extract('(\+\d+)')
df['cell'] = df['cell'].str.extract('(\+\d+)')
# Checking the updated DataFrame
print("Question 2:")
df.shape  
df.head(25)
```

# QUESTION 3: Load
Load the updated DataFrame from Question 2 into a Postgres table named user_accounts
Query the new postgres table in python and display the output results
Use pd.read_sql() to query your new table and verify it’s been loaded correctly


```python
import psycopg2

# Establish a connection to the PostgreSQL database
conn = psycopg2.connect(host='localhost', port='5432', database='gc', user='postgres', password='welcome123')

# Create a cursor object
cur = conn.cursor()

# Create the user_accounts table
cur.execute('''
    CREATE TABLE user_accounts (
        gender VARCHAR(10),
        name_title VARCHAR(10),
        name_first VARCHAR(255),
        name_last VARCHAR(255),
        location_street VARCHAR(255),
        location_city VARCHAR(255),
        location_state VARCHAR(255),
        location_country VARCHAR(255),
        location_postcode INTEGER,
        email VARCHAR(255),
        login_username VARCHAR(255),
        login_password VARCHAR(255),
        login_salt VARCHAR(255),
        dob_date DATE,
        dob_age INTEGER,
        registered_date DATE,
        registered_age INTEGER,
        phone VARCHAR(255),
        cell VARCHAR(255),
        id_name VARCHAR(255),
        id_value VARCHAR(255),
        nat VARCHAR(10),
        picture_large VARCHAR(255),
        picture_medium VARCHAR(255),
        picture_thumbnail VARCHAR(255),
        na_user BOOLEAN,
        age_at_registration INTEGER
    )
''')

# Commit the changes
conn.commit()

# Close the cursor and the connection
cur.close()
conn.close()
```
